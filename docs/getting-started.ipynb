## The Jupyter Environment

``roboworld`` was developed to be used within the [Jupyter notebook environment](https://jupyter.org/). 
Notebooks allow you to display the current state of the ``World`` and its ``Robo`` via a matrix plot generated by ``matplotlib``. 
A recorder ``ArrayRecorder`` saves a frame (basically a matrix) everytime ``Robo`` changes its or the world’s state. Furthermore, it is possible to display an animation of the executed instructions within a Jupyter notebook.

The direct visual feedback is an important feature of roboworld and ment to improve the student’s learning process. Therefore, we recommand you to use Jupyter notebook. Follow the [installaiton instructions](https://jupyter.org/install).

## The Student's Workflow
Students should recieve a specific ``World`` and a description what to do with it. For example:

**Exercise 1.** Given is the following ``World``. Move ``robo`` from left to its goal in the east. 

*Hint*: ``robo`` is oriented towards the east.

````python
import roboworld as rw
````

````python
world = rw.corridor()
robo = world.robo
````

````python
world.show()
````

**Solution 1.**

````python
while not robo.is_at_goal():
    robo.move()
````
